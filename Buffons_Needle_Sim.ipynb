## Buffon's needle 

The problem is to drop a needle on the unit square and count the number of times that the needle touches (or "cuts") the edge of the square. This is a classic problem, but instead of going directly for the analytical result, our approach here is to develop a quick simulation and then refine the simulation to obtain more accurate results more efficiently. This kind of reasoning is common in Monte Carlo methods and we will work through this example in careful detail.

Even though this is a classic problem, it is easy to understand and has enough complexity characteristic of much more difficult problems that one encounters in practice. The overall idea is that the problem is too complex to analyze analytically (at least in the general case), so we want to use our computational power to mock up the problem in a simulation and then derive our solution for our simulation.

The purpose of this section is to connect a coded simulation to the underlying mathematical reasoning and then use this reasoning to improve the simulation.

```python
from __future__ import  division
from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle
```

This sets up the random needle position and orientation and creates the corresponding matplotlib primitives to be drawn in the figure.

```python
def needle(L=0.5):
    '''
    Drops needle on unit square with associated matplotlib primitives.
    
    L:= length of needle (default=0.5)
    '''
    xc,yc = rand(2) 
    ang = rand()*pi
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1):
        #all contained in square
        needle.blue_count+=1
        return ([Line2D([x0,xe],[y0,ye],alpha=0.3) , Line2D([xc],[yc],marker='o',alpha=0.3)] )
    else:
        needle.red_count+=1
        return ([Line2D([x0,xe],[y0,ye],color='r',alpha=0.3), Line2D([xc],[yc],marker='o',color='r')] )
```

Now, we can simulate the dropping event on the square (yellow background ). The red needles cut the edge of the square and the blue ones are completely contained therein. The circle marker shows the needle position.

```python
fig,ax=subplots()
fig.set_size_inches(5,5)

needle.red_count=0 # reset counters
needle.blue_count=0
for k in range(500):
    for i in needle(.5): ax.add_line(i)

ax.set_aspect(1)
ax.add_patch(Rectangle((0,0),1,1,alpha=0.3,color='y'))
ax.set_xlabel('x-direction')
ax.set_ylabel('y-direction')
ax.axis([-.1,1.1,-.1,1.1])
ax.set_title(r'$\mathbb{P}$ (touch square)=%3.2f'% (needle.red_count/(needle.blue_count+needle.red_count)))
ax.grid()
```

The figure we just drew is a good visual check of our simulation and helps motivate our reasoning.  We'll need this next  code for later, but it's basically just another version of the previous graph with different return values and internals.

```python
def needle_gen(L=0.5,return_data=False):
    '''ang := angle orientation of needle in radians
    L:= length of needle
    '''
    assert 0<L<1
    xc,yc = rand(2)
    ang = rand()*pi
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1): # no cut
        if return_data:
            return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=False)
        else:
            return False 
    else: # cut
        if return_data:
            return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=True)
        else:
            return True 
```

## Formal Analysis

The needle is entirely contained in the square ( at least in the x-direction) when the following two conditions hold:

$$ 0 < x_c - L/2 \cos(\theta) < 1 $$

$$ 0 < x_c + L/2 \cos(\theta) < 1 $$

combining these two into one big inequality gives:

$$ \max(-L/2 \cos(\theta) , L/2 \cos(\theta) ) <  x_c < \min(1-L/2 \cos(\theta) , 1+L/2 \cos(\theta) )$$.

In the case where $0< \theta < \pi/2 $, we additionally have the following:

$$  L/2 \cos(\theta)  <  x_c < 1-L/2 \cos(\theta) $$

Because $\theta$ is uniformly distributed between $0$ and $\pi$, for this restriction domain of $\theta$ and $x_c$, we have the probability of the needle being contained completely in the square as the following:

$$ \frac{1}{\pi}\int_0^{\pi/2} \int_{L/2 \cos(\theta)}^{1-L/2 \cos(\theta)} dx_c d\theta = \frac{1}{\pi}\int_0^{\pi/2} 1-L \cos(\theta) d\theta = \frac{1}{2} - \frac{L}{\pi} $$ 

Now, we can consider the other half of the domain of $\theta$, $ \pi/2 < \theta < \pi $, where we have

$$  -L/2 \cos(\theta)  <  x_c < 1+L/2 \cos(\theta)$$

Then, following the same reasoning as before we obtain:

$$ \frac{1}{\pi}\int_{\pi/2}^{\pi} \int_{-L/2 \cos(\theta)}^{1+L/2 \cos(\theta)} dx_c d\theta = \frac{1}{\pi}\int_{\pi/2}^{\pi} 1+L \cos(\theta) d\theta = \frac{1}{2} - \frac{L}{\pi} $$ 

Thus, combining these two results gives the probability of the needle **not**  cutting the edge in the x-direction as the following:

$$ \mathbb{P}(\text{no x-cut}) = 1 - \frac{2 L}{\pi}$$

Now, we can pursue exactly the same line of reasoning for the y-direction, or, just recognize that by symmetry it would be exactly the same as for the x-direction.  Nonetheless, let's be thorough and reproduce the reasoning for the y-direction, just to be on the safe side.

As before, the condition in the y-direction for the needle being entirely contained in the square is the following:

$$ \max(-L/2 \sin(\theta) , L/2 \sin(\theta) ) <  y_c < \min(1-L/2 \sin(\theta) , 1+L/2 \sin(\theta) )$$

which can be combined into one big inequality, given the restrictions on $\theta \in (0,\pi)$ as shown:

$$  L/2 \sin(\theta) <  y_c < 1-L/2 \sin(\theta) $$

Then, integrating this out as before, we obtain

$$ \frac{1}{\pi}\int_0^{\pi} \int_{L/2 \sin(\theta)}^{1-L/2 \sin(\theta)} dy_c d\theta = \frac{1}{\pi}\int_0^{\pi} 1-L \sin(\theta) d\theta = 1 - \frac{2 L}{\pi} $$ 

$$ \mathbb{P}(\text{no y-cut}) = 1 - \frac{2 L}{\pi}$$

which is exactly what we got for the x-direction. It's usually worth the extra work to check things like this instead of trying to be too clever too early.

Then, combining the x-direction and y-direction results, we obtain,

$$ \mathbb{P}(\text{no cut}) = \left(1 - \frac{2 L}{\pi}\right)^2$$

so that

$$ \mathbb{P}(\text{cut}) =1-\mathbb{P}(\text{no cut}) = 1-\left(1 - \frac{2 L}{\pi}\right)^2$$

which we code next.

```python
prob_cut= lambda l:(1-(1-2*l/pi)**2)
```

The next code-block setups a the same simulation as before, except now we can compare our estimated probability with the theoretical result we just discovered.

```python
L = .5
fig,axs = subplots(1,2)
fig.set_size_inches((12,2))
ax=axs[0]
samples=np.array([[needle_gen(L) for i in range(300)] for k in range(100)])
errors=samples - prob_cut(L)
ax.plot(np.mean(errors,axis=1),'o-',alpha=0.3)
ax.axis(xmax=errors.shape[0])
ax.set_xlabel(r'$trial$',fontsize=18)
ax.set_ylabel('$\mathbb{\hat{P}}(cut)$',fontsize=18)
ax.set_title('error = %3.3f, std=%3.3f'%(errors.mean(),(errors.mean(1)).std()))
ax.grid()

ax = axs[1]
ax.hist( errors.mean(1));
```

## Using weighting for better efficiency

There's nothing wrong with the way we have been using the simulation so far, but in the 1st figure all of those blue needles did not help us compute the probability we were after because they did not touch the edges of the square. Yes, we could reverse the reasoning and use the blue instead of the red for our estimation, but that does not escape the basic problem. For example, if the length of the needle were really short in comparison, then there would have been a lot of blue in our initial plot. In the extreme case, we could find ourselves in a situation where we run the simulation for a long time and not have a single cut! We need to be more efficient, if possible, and  we want to use our computer time to create samples that will reduce our sample history variance.

Let's get more explicit. We computed a matrix of (100x300) cases of needle drops. Each of element of `samples` in the code is a True/False result as to whether or not the needle touched the edge of the square. We averaged over the 300 rows to compute the estimated probability of a cut (the mean value of the boolean array in this case).  The estimated standard deviation is taken over the ensemble of these estimated means. 

A common way to evaluate the quality of the simulated result is to compute the relative error, which is the ratio of the estimated standard deviation of the mean to the estimated mean using *all* of the samples. This is computed in the title of the figure below.

```python
fig,ax=subplots()
fig.set_size_inches(12,2)
ax.plot(samples.mean(1))
ax.set_title(r'$\hat{\sigma}=%2.3f,R_e=%2.3f$'% (samples.mean(1).std(),samples.mean(1).std()/samples.mean()), fontsize=18)
ax.set_xlabel('sample history index')
```

So far, our relative error is below the 5% rule of thumb, but let's see how we do when the needle length is smaller? The code-block below draws the corresponding figure for the simulation for this case.

```python
L=0.1

samples=np.array([[needle_gen(L) for i in range(300)] for k in range(100)])

fig,ax=subplots()
fig.set_size_inches(12,2)
ax.plot(samples.mean(1))
ax.set_title(r'$\hat{\sigma}=%2.3f,R_e=%2.3f$'% (samples.mean(1).std(),samples.mean(1).std()/samples.mean()), fontsize=18)
ax.set_xlabel('sample history index')
```

The figure above shows that for shorter needles, we did not do so well for relative error. Let's see if we can create more useful samples in this case. This situation is illustrated in the figure below.

```python
fig,ax=subplots()
fig.set_size_inches(5,5)

needle.red_count=0 # reset counters
needle.blue_count=0
for k in range(500):
    for i in needle(.1): ax.add_line(i)

ax.set_aspect(1)
ax.add_patch(Rectangle((0,0),1,1,alpha=0.3,color='y'))
ax.set_xlabel('x-direction')
ax.set_ylabel('y-direction')
ax.axis([-.1,1.1,-.1,1.1])
ax.set_title(r'$\mathbb{P}$ (touch square)=%3.2f'% (needle.red_count/(needle.blue_count+needle.red_count)))
ax.grid()
```

The figure shows there are very few cuts in our simulation (red needles). To remedy this, we can alter the code to bias for randomly generating x-coordinates that are within needle length of an edge instead of uniformly random in the unit interval. 

```python
def needle_b(L=0.5):
    ''' biased towards x-coordinate edges
    ang := angle orientation of needle in radians
    L:= length of needle
    '''
    yc = rand() # uniform on unit interval
    r = L*rand()/2.
    xc = r if rand()<0.5 else (1-r) # pick either edge 50/50
    ang = rand()*pi
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1):
        #all contained in square
        needle.blue_count+=1
        return ([Line2D([x0,xe],[y0,ye],alpha=0.3) , Line2D([xc],[yc],marker='o',alpha=0.3)] )
    else:
        needle.red_count+=1
        return ([Line2D([x0,xe],[y0,ye],color='r',alpha=0.3), Line2D([xc],[yc],marker='o',color='r')] )
```

```python
fig,ax=subplots()
fig.set_size_inches(5,5)

needle.red_count=0 # reset counters
needle.blue_count=0
for k in range(500):
    for i in needle_b(L): ax.add_line(i)

ax.set_aspect(1)
ax.add_patch(Rectangle((0,0),1,1,alpha=0.3,color='y'))
ax.set_xlabel('x-direction')
ax.set_ylabel('y-direction')
ax.axis([-.1,1.1,-.1,1.1])
ax.set_title(r'$\mathbb{P}$ (touch square)=%3.2f'% (needle.red_count/(needle.blue_count+needle.red_count)))
ax.grid()
```

Now, we can likewise change the other code-block to reflect this bias for the edges.

```python
def needle_gen_b(L=0.5,return_data=False):
    '''ang := angle orientation of needle in radians
    L:= length of needle
    '''
    assert 0<L<1
    yc = rand() # uniform on unit interval
    r = L*rand()/2. # uniform between in [0,L]
    xc = r if rand()<0.5 else (1-r) # pick either edge 50/50
    ang = rand()*pi # uniform between [0,pi]
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1): # no cut
        if return_data:
            return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=False)
        else:
            return False 
    else: # cut
        if return_data:
            return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=True)
        else:
            return True 
```

Now, let's crank through a quick simulation to see how this works. Notice how for the same number of samples we have a **much** better relative error.

```python
L=0.1
samples_b=np.array([[needle_gen_b(L) for i in range(300)] for k in range(100)])

fig,ax=subplots()
fig.set_size_inches(12,2)
ax.plot(samples_b.mean(1))
ax.set_title(r'$\hat{\sigma}=%2.3f,R_e=%2.3f$'% (samples_b.mean(1).std(),samples_b.mean(1).std()/samples_b.mean()), fontsize=18)
ax.set_xlabel('sample history index',fontsize=16);
```

Now, all we have to do is correct for our bias. The probability of being near the edges  is $2 L$, we have to multiply our estimated probability by this factor
Note that the spread of the estimated means is much tighter in the biased case.

```python
fig,ax=subplots()
fig.set_size_inches(10,3)
ax.hist(samples.mean(1),alpha=0.3,normed=1,label='unbiased')
ax.hist(samples_b.mean(1)*2*L,alpha=0.3,normed=1,label='biased')
ax.legend(loc=0)
ax.set_title('Biased samples more efficiently produce useful samples')
```
