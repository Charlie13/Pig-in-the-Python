```python
from __future__ import  division
from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle
```

```python
def needle(L=0.5):
    '''ang := angle orientation of needle in radians
    L:= length of needle
    '''
    xc,yc = rand(2) 
    ang = rand()*pi
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1):
        #all contained in square
        needle.blue_count+=1
        return ([Line2D([x0,xe],[y0,ye],alpha=0.3) , Line2D([xc],[yc],marker='o',alpha=0.3)] )
    else:
        needle.red_count+=1
        return ([Line2D([x0,xe],[y0,ye],color='r',alpha=0.3) , Line2D([xc],[yc],marker='o',color='r')] )
```

```python
fig,ax=subplots()
needle.red_count=0 # reset counters
needle.blue_count=0
for k in range(500):
    for i in needle(.5): ax.add_line(i)

ax.set_aspect(1)
ax.add_patch(Rectangle((0,0),1,1,alpha=0.3,color='y'))
ax.axis([-.1,1.1,-.1,1.1])
ax.set_title('Prob touch square=%3.3f'% (needle.red_count/(needle.blue_count+needle.red_count)))
ax.grid()
```

```python
def needle_gen(L=0.5,return_data=False):
    '''ang := angle orientation of needle in radians
    L:= length of needle
    '''
    assert 0<L<1
    xc,yc = rand(2)
    ang = rand()*pi
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1): # no cut
        if return_data:
            return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=False)
        else:
            return False 
    else: # cut
        if return_data:
            return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=True)
        else:
            return True 
```

$$ 0 < x_c - L/2 \cos(\theta) < 1 $$

$$ 0 < x_c + L/2 \cos(\theta) < 1 $$

$$ \max(-L/2 \cos(\theta) , L/2 \cos(\theta) ) <  x_c < \min(1-L/2 \cos(\theta) , 1+L/2 \cos(\theta) )$$

Given $ 0< \theta < \pi/2 $, we have

$$  L/2 \cos(\theta)  <  x_c < 1-L/2 \cos(\theta)$$

Then,

$$ \frac{1}{\pi}\int_0^{\pi/2} \int_{L/2 \cos(\theta)}^{1-L/2 \cos(\theta)} dx_c d\theta = \frac{1}{\pi}\int_0^{\pi/2} 1-L \cos(\theta) d\theta = \frac{1}{2} - \frac{L}{\pi} $$ 

Given $ \pi/2 < \theta < \pi $, we have

$$  -L/2 \cos(\theta)  <  x_c < 1+L/2 \cos(\theta)$$

Then,

$$ \frac{1}{\pi}\int_{\pi/2}^{\pi} \int_{-L/2 \cos(\theta)}^{1+L/2 \cos(\theta)} dx_c d\theta = \frac{1}{\pi}\int_{\pi/2}^{\pi} 1+L \cos(\theta) d\theta = \frac{1}{2} - \frac{L}{\pi} $$ 

Thus,

$$ \mathbb{P}(\text{no x-cut}) = 1 - \frac{2 L}{\pi}$$

$$ \max(-L/2 \sin(\theta) , L/2 \sin(\theta) ) <  y_c < \min(1-L/2 \sin(\theta) , 1+L/2 \sin(\theta) )$$

$$  L/2 \sin(\theta) <  y_c < 1-L/2 \sin(\theta) $$

$$ 1-L \sin(\theta)  $$

Given $ 0< \theta < \pi $, we have

$$  L/2 \sin(\theta)  <  y_c < 1-L/2 \sin(\theta)$$

Then,

$$ \frac{1}{\pi}\int_0^{\pi} \int_{L/2 \sin(\theta)}^{1-L/2 \sin(\theta)} dy_c d\theta = \frac{1}{\pi}\int_0^{\pi} 1-L \sin(\theta) d\theta = 1 - \frac{2 L}{\pi} $$ 

$$ \mathbb{P}(\text{no y-cut}) = 1 - \frac{2 L}{\pi}$$

Then,

$$ \mathbb{P}(\text{no cut}) = \left(1 - \frac{2 L}{\pi}\right)^2$$

so that

$$ \mathbb{P}(\text{cut}) =1-\mathbb{P}(\text{no cut}) = 1-\left(1 - \frac{2 L}{\pi}\right)^2$$



```python
prob_cut= lambda l:(1-(1-2*l/pi)**2)
```

```python
L = .5
fig,axs = subplots(1,2)
fig.set_size_inches((10,3))
ax=axs[0]
errors=np.array([[needle_gen(L) for i in range(300)] for k in range(100)])-prob_cut(L)
ax.plot(np.mean(errors,axis=1),'o-',alpha=0.3)
ax.axis(xmax=errors.shape[0])
ax.set_xlabel('$trial$',fontsize=18)
ax.set_ylabel('$\mathbb{\hat{P}}(cut)$',fontsize=18)
ax.set_title('error = %3.3f, std=%3.3f'%(errors.mean(),(errors.mean(1)).std()))
ax.grid()

ax = axs[1]
ax.hist( errors.mean(1));
```

```python
import pandas as pd

df=pd.DataFrame([needle_gen(return_data=1) for i in range(300*100)])

def gb(v,df=df):
    return df.groupby('cut').get_group(v)[['xc','yc']].values.T 

fig,ax = subplots()
ax.set_aspect(1)
xc,yc=gb(True)
ax.plot(xc,yc,'ro',alpha=.5)
xc,yc=gb(False)
ax.plot(xc,yc,'bo',alpha=0.3)

xc,yc,cut=df[(df.xc > L/2) & (df.xc < 1-L/2) & ( df.yc < 1-L/2) & (df.yc>L/2)][['xc','yc','cut']].values.T
assert sum(cut)==0 # no cuts possible in this square
ax.plot(xc,yc,'go',alpha=0.3)
```

```python
def favor_gen(L=0.5):
    assert 0<L<1
    if randint(0,2):
        xc= (rand()*L/2 , rand()*L/2+(1-L/2))[randint(0,2)]
        yc= rand()
    else:
        xc= rand()
        yc= (rand()*L/2 , rand()*L/2+(1-L/2))[randint(0,2)]
    ang = rand()*pi
    x0,y0 = xc-L/2*cos(ang),yc-L/2*sin(ang)
    xe,ye = xc+L/2*cos(ang),yc+L/2*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1): # no cut
        return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=False)
    else: # cut
        return dict(x0=x0,y0=y0,xc=xc,yc=yc,xe=xe,ye=ye,ang=ang,cut=True)
```

```python
fd=pd.DataFrame([favor_gen() for i in range(300*100)])
errors=prob_cut(L)-(fd.cut.values.reshape(100,-1).astype(int)) * (1-(1-L)**2)
#errors=(fd.cut.values.reshape(-1,100).astype(int)) * (1-(1-L)**2)
```

```python
fig,axs = subplots(1,2)
fig.set_size_inches((10,3))

ax=axs[0]
ax.plot(np.mean(errors,axis=1),'o-',alpha=0.3)
ax.axis(xmax=errors.shape[0])
ax.set_xlabel('$trial$',fontsize=18)
ax.set_ylabel('$\mathbb{\hat{P}}(cut)$',fontsize=18)
ax.set_title('error = %3.3f, std=%3.3f'%(errors.mean(),(errors.mean(1)).std()))
ax.grid()

ax = axs[1]
ax.hist(errors.mean(1));
```
